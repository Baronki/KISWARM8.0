## Step 4: Health Check

```python
# ═══════════════════════════════════════════════════════════════
# KISWARM7 - SYSTEM HEALTH CHECK
# ═══════════════════════════════════════════════════════════════

import json
from datetime import datetime

def check_kiswarm_health():
    """Run a comprehensive health check on KISWARM7 setup."""
    
    results = {
        "timestamp": datetime.now().isoformat(),
        "status": "OPERATIONAL",
        "version": "7.0",
        "checks": []
    }
    
    # Check 1: Repository
    repo_exists = os.path.exists('/content/KISWARM7')
    results["checks"].append({
        "name": "Repository Clone",
        "status": "PASS" if repo_exists else "FAIL",
        "detail": "/content/KISWARM7 exists" if repo_exists else "Repository not found"
    })
    
    # Check 2: Backend Directory
    backend_exists = os.path.exists('/content/KISWARM7/backend/python')
    results["checks"].append({
        "name": "Backend Directory",
        "status": "PASS" if backend_exists else "FAIL",
        "detail": "backend/python exists" if backend_exists else "backend/python not found"
    })
    
    # Check 3: Sentinel Modules
    sentinel_path = '/content/KISWARM7/backend/python/sentinel'
    sentinel_count = 0
    if os.path.exists(sentinel_path):
        sentinel_count = len([f for f in os.listdir(sentinel_path) if f.endswith('.py')])
    results["checks"].append({
        "name": "Sentinel Modules",
        "status": "PASS" if sentinel_count > 50 else "WARN",
        "detail": f"{sentinel_count} Python files found"
    })
    
    # Check 4: KIBank Modules
    kibank_path = '/content/KISWARM7/backend/python/kibank'
    kibank_count = 0
    if os.path.exists(kibank_path):
        kibank_count = len([f for f in os.listdir(kibank_path) if f.endswith('.py')])
    results["checks"].append({
        "name": "KIBank Modules",
        "status": "PASS" if kibank_count > 15 else "WARN",
        "detail": f"{kibank_count} Python files found"
    })
    
    # Check 5: Python Path
    pythonpath_ok = '/content/KISWARM7/backend' in sys.path
    results["checks"].append({
        "name": "Python Path",
        "status": "PASS" if pythonpath_ok else "FAIL",
        "detail": "PYTHONPATH correctly configured" if pythonpath_ok else "PYTHONPATH not set"
    })
    
    # Check 6: Core Dependencies
    deps_ok = True
    missing_deps = []
    for dep in ['flask', 'requests', 'rich']:
        try:
            __import__(dep)
        except ImportError:
            deps_ok = False
            missing_deps.append(dep)
    results["checks"].append({
        "name": "Core Dependencies",
        "status": "PASS" if deps_ok else "FAIL",
        "detail": "All core deps available" if deps_ok else f"Missing: {missing_deps}"
    })
    
    # Calculate overall status
    failed = sum(1 for c in results["checks"] if c["status"] == "FAIL")
    if failed > 0:
        results["status"] = "DEGRADED"
    
    return results

# Run health check
health = check_kiswarm_health()

# Display results
from rich.console import Console
from rich.table import Table

console = Console()

console.print("\n")
console.print("╔═══════════════════════════════════════════════════════════════╗", style="bold blue")
console.print("║            KISWARM7 - HEALTH CHECK RESULTS                    ║", style="bold blue")
console.print("╚═══════════════════════════════════════════════════════════════╝", style="bold blue")

table = Table(title="\nSystem Status")
table.add_column("Check", style="cyan")
table.add_column("Status", style="bold")
table.add_column("Detail")

for check in health["checks"]:
    status_style = "green" if check["status"] == "PASS" else "yellow" if check["status"] == "WARN" else "red"
    table.add_row(check["name"], f"[{status_style}]{check['status']}[/{status_style}]", check["detail"])

console.print(table)

console.print(f"\n📊 Overall Status: [bold {'green' if health['status'] == 'OPERATIONAL' else 'yellow'}]{health['status']}[/bold {'green' if health['status'] == 'OPERATIONAL' else 'yellow'}]")
console.print(f"📅 Timestamp: {health['timestamp']}\n")
```

## Step 5: Start Minimal API Server

```python
# ═══════════════════════════════════════════════════════════════
# KISWARM7 - START MINIMAL API SERVER
# ═══════════════════════════════════════════════════════════════

from flask import Flask, jsonify
from flask_cors import CORS
import threading
import time

app = Flask(__name__)
CORS(app)

@app.route('/')
def index():
    return jsonify({
        "name": "KISWARM7",
        "version": "7.0",
        "status": "OPERATIONAL",
        "modules": 83,
        "ki_agents": 27,
        "docker_free": True,
        "repository": "https://github.com/Baronki/KISWARM7"
    })

@app.route('/health')
def health():
    return jsonify({
        "status": "HEALTHY",
        "timestamp": datetime.now().isoformat(),
        "checks": {
            "repository": "OK",
            "backend": "OK",
            "dependencies": "OK"
        }
    })

@app.route('/api/status')
def api_status():
    return jsonify({
        "system": "KISWARM7",
        "version": "7.0-NATIVE",
        "environment": "Google Colab",
        "modules": {
            "total": 83,
            "sentinel": 61,
            "kibank": 19,
            "mesh": 7,
            "cognitive": 4
        }
    })

# Start server in background
def run_server():
    app.run(host='0.0.0.0', port=5002, use_reloader=False)

server_thread = threading.Thread(target=run_server, daemon=True)
server_thread.start()

print("✅ KISWARM7 API Server started on port 5002")
print("📡 Endpoints:")
print("   - http://localhost:5002/")
print("   - http://localhost:5002/health")
print("   - http://localhost:5002/api/status")
```

## Step 6: Create Public Tunnel (ngrok)

```python
# ═══════════════════════════════════════════════════════════════
# KISWARM7 - CREATE PUBLIC TUNNEL
# ═══════════════════════════════════════════════════════════════

# Install pyngrok if needed
!pip install -q pyngrok

from pyngrok import ngrok
import requests

# Create public tunnel
public_url = ngrok.connect(5002)
print(f"🚀 Public URL: {public_url}")

# Test the connection
import time
time.sleep(2)  # Wait for tunnel to be ready

try:
    response = requests.get(
        f"{public_url}/health",
        headers={"ngrok-skip-browser-warning": "true"}
    )
    print(f"\n✅ Tunnel test: {response.status_code}")
    print(f"📦 Response: {response.json()}")
except Exception as e:
    print(f"⚠️ Tunnel test failed: {e}")
```
